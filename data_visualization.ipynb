## Language

```python
# visulize the language column using pie chart
plt.figure()
eng_data['Language'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
import matplotlib.pyplot as plt

from wordcloud import WordCloud, STOPWORDS
plt.figure(figsize=(12,12))
stopwords = set(STOPWORDS)
wordcloud = WordCloud(background_color='white',width=1000,height=1000,stopwords=stopwords).generate(str(eng_data['Language']))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
# visulize the language column using cloud word
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
text = eng_data['Language'].values
wordcloud = WordCloud(
    width = 3000,
    height = 2000,
    background_color = 'black',
    stopwords = STOPWORDS).generate(str(text))
fig = plt.figure(
    figsize = (20, 20),
    facecolor = 'k',
    edgecolor = 'k')
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```

## instgram dataset

```python
insta_data = eng_data.drop(['Threads account', 'Threads URL',
       'Threads Follower #', 'YouTube account', 'YouTube URL',
       'YouTube Subscriber #', 'TikTok account', 'TikTok URL','X (Twitter) handle',
         'X (Twitter) URL', 'X (Twitter) Follower #',
       'Facebook page', 'Facebook URL', 'Facebook Follower #',
       'TikTok Subscriber #'],axis=1)
print(insta_data.columns)
insta_data = insta_data.dropna().reset_index(drop=True)
#save the data to a csv file
insta_data.to_csv('insta_data.csv')
```

## facebook dataset

```python
fb_data = eng_data.drop(['Threads account', 'Threads URL',
       'Threads Follower #', 'YouTube account', 'YouTube URL',
       'YouTube Subscriber #', 'TikTok account', 'TikTok URL','X (Twitter) handle',
         'X (Twitter) URL', 'X (Twitter) Follower #' ,'Instragram page',
       'Instagram URL', 'Instagram Follower #',
       'TikTok Subscriber #'],axis=1)
print(fb_data.columns)
fb_data = fb_data.dropna().reset_index(drop=True)
#save the data to a csv file
fb_data.to_csv('fb_data.csv')
```

## youtube dataset

```python
youtube_data = eng_data.drop(['Threads account', 'Threads URL',
       'Threads Follower #', 'TikTok account', 'TikTok URL','X (Twitter) handle',
         'X (Twitter) URL', 'X (Twitter) Follower #' ,'Instragram page', 'Facebook page',
       'Facebook URL', 'Facebook Follower #',
       'Instagram URL', 'Instagram Follower #',
       'TikTok Subscriber #'],axis=1)
print(youtube_data.columns)
youtube_data = youtube_data.dropna().reset_index(drop=True)
#save the data to a csv file
youtube_data.to_csv('youtube_data.csv')
```

## twitter dataset

```python
twtr_data = eng_data.drop(['Threads account', 'Threads URL',
       'Threads Follower #', 'TikTok account', 'TikTok URL','Instragram page', 'Facebook page',
       'Facebook URL', 'Facebook Follower #',
       'Instagram URL', 'Instagram Follower #', 'YouTube account',
       'YouTube URL', 'YouTube Subscriber #',
       'TikTok Subscriber #'],axis=1)
print(twtr_data.columns)
twtr_data = twtr_data.dropna().reset_index(drop=True)
#save the data to a csv file
twtr_data.to_csv('twtr_data.csv')
```

## TikTok dataset

```python
tktok_data = eng_data.drop(['Threads account', 'Threads URL',
       'Threads Follower #','Instragram page', 'Facebook page',
       'Facebook URL', 'Facebook Follower #',
       'Instagram URL', 'Instagram Follower #', 'YouTube account',
       'YouTube URL', 'YouTube Subscriber #','X (Twitter) handle', 'X (Twitter) URL', 'X (Twitter) Follower #'],axis=1)
print(tktok_data.columns)
tktok_data = tktok_data.dropna().reset_index(drop=True)
#save the data to a csv file
tktok_data.to_csv('tktok_data.csv')
```

## Thread dataset

```python
thrd_data = eng_data.drop(['Instragram page', 'Facebook page',
       'Facebook URL', 'Facebook Follower #',
       'Instagram URL', 'Instagram Follower #', 'YouTube account',
       'YouTube URL', 'YouTube Subscriber #','X (Twitter) handle', 'X (Twitter) URL', 'X (Twitter) Follower #', 'TikTok account',
       'TikTok URL', 'TikTok Subscriber #'],axis=1)
print(thrd_data.columns)
thrd_data = thrd_data.dropna().reset_index(drop=True)
#save the data to a csv file
thrd_data.to_csv('thrd_data.csv')
```

### heighest language subscreibers

```python
#subplt for all platforms
fig, ax = plt.subplots(2,3,figsize=(20,10))
fig.suptitle('Top 10 languages with highest followers')
#insta
insta_data.groupby('Language')['Instagram Follower #'].max().sort_values(ascending=False).head(10).plot.bar(ax=ax[0,0])
ax[0,0].set_title('Instagram')
ax[0,0].set_xlabel('Language')
ax[0,0].set_ylabel('Instagram Follower #')
#facebook
fb_data.groupby('Language')['Facebook Follower #'].max().sort_values(ascending=False).head(10).plot.bar(ax=ax[0,1])
ax[0,1].set_title('Facebook')
ax[0,1].set_xlabel('Language')
ax[0,1].set_ylabel('Facebook Follower #')
#youtube
youtube_data.groupby('Language')['YouTube Subscriber #'].max().sort_values(ascending=False).head(10).plot.bar(ax=ax[0,2])
ax[0,2].set_title('YouTube')
ax[0,2].set_xlabel('Language')
ax[0,2].set_ylabel('YouTube Subscriber #')
#twitter
twtr_data.groupby('Language')['X (Twitter) Follower #'].max().sort_values(ascending=False).head(10).plot.bar(ax=ax[1,0])
ax[1,0].set_title('Twitter')
ax[1,0].set_xlabel('Language')
ax[1,0].set_ylabel('Twitter Follower #')
#threads
thrd_data.groupby('Language')['Threads Follower #'].max().sort_values(ascending=False).head(10).plot.bar(ax=ax[1,1])
ax[1,1].set_title('Threads')
ax[1,1].set_xlabel('Language')
ax[1,1].set_ylabel('Threads Follower #')
#tiktok
tktok_data.groupby('Language')['TikTok Subscriber #'].max().sort_values(ascending=False).head(10).plot.bar(ax=ax[1,2])
ax[1,2].set_title('TikTok')
ax[1,2].set_xlabel('Language')
ax[1,2].set_ylabel('TikTok Subscriber #')
plt.subplots_adjust(hspace=0.8)
plt.show()
```
